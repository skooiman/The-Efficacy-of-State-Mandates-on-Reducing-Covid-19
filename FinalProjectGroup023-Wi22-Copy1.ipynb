# Data Cleaning

The first step in our data cleaning process was to merge the nyt infection/deaths dataframe with the state policy dataframe.

This was done by creating a function that

- Drops the fips column as it was an unecassary column that was used to numerically label states

- Creates a subdataframe consisting of all the infections and deaths that were given for a specific state that is named in the function call.

- Creates a separate df for the deaths on a specific date in the state

- Creates a separate df for the cases on a specific date in the state

- Merges the two cases and deaths dataframes together

- Reorders the rows of the new dataframe so the cases row and deaths row of the same date are       next to each other.

- The date is set to be the index

- The resulting df is transposed so that the state is now the row of the dataframe, and the      
  columns are the number of deaths on a certain date and the number of cases on a certain date.

```python
#Function that takes in a dataframe and a named state and converts it
#to a data frame that will merge with the COVID-19 US State Policy Database

def convert_state_data(state ,df):
    
    import pandas as pd
    import warnings
    warnings.filterwarnings("ignore")
   
    #drop the unecassary fip num
    df1 =df.drop(["fips"],axis =1)
    
    #pull all the files that have the requested state
    Statedf = df1.loc[df1['state'] == state]
    
    #add deaths to the end of date string and create df of deaths in state
    Statedf2 = Statedf['date'] = Statedf['date'].astype(str) + " deaths"
    Statedeaths = Statedf.drop(["cases"],axis =1)
    Statedeaths2 =Statedeaths.rename(columns={"deaths": state})
    
    
    #add cases to the end of date string and create df of cases in state
    Statedf = df1.loc[df1['state'] == state]
    Statedf2 = Statedf['date'] = Statedf['date'].astype(str) + " cases"
    Statecases = Statedf.drop(["deaths"],axis =1)
    Statecases2 = Statecases.rename(columns={"cases": state})
    
    #merge the cases and deaths dataframes
    frames = [Statedeaths2, Statecases2]
    result = pd.concat(frames)
    df100 =pd.DataFrame(result)
    
    #resort to line up deaths and cases by date
    df101 = df100.sort_values(by=['date'])
    
    # drop the state col and set date as index
    df102 = df101.drop(["state"],axis =1)
    df103 = df102.set_index("date")
    
    #convert rows to col and col to rows
    df104 = df103.transpose()
    
    
    return df104
```

The following for-loop uses the convert_state_data function and the list of 50 states to convert the nyt/covid-19 df to one that can be merged with the US state policy df. It does this by applying the function on each state on the list and then appending the results to a newdf. The State column is then reset to be the index.
Finally all columns that have more than 50% null values for the covid 19 cases and deaths are dropped.These null values ae present because some states did not update their data daily.
They were dropped becasue there is not information in those columns to help with analysis.

```python
#For loop to apply convert_state function to nyt database using list of every state
newdf = pd.DataFrame()
for regions in states:
    added_state =convert_state_data(regions, df)
    newdf = newdf.append(added_state)
    
#reset index to separate the state names then rename the column
newdf.reset_index(inplace=True)    
newdf.rename(columns={"index": "STATE"}, inplace=True)
    
#drop dates with data missing for more than half the states since that's not enough to compare
newdf.dropna(axis='columns', thresh=25, inplace=True)
    
#newdf data
newdf.head()
```

This data in this data frame is a cumulative count of cases and deaths. However, in order to analyze our research question a daily count of NEW cases and deaths per day is needed. 
The following code creates two subdataframes, one for cases and one for deaths that contains a daily count of new cases and deaths in each state and not the running total for each state.

```python
cases_data = {}
deaths_data = {}
cases_data['STATE'] = newdf["STATE"]
deaths_data['STATE'] = newdf["STATE"]
for col in newdf.columns:
    if "cases" in col:
        cases_data[col] = newdf[col]
    elif "deaths" in col:
        deaths_data[col] = newdf[col]
cases = pd.DataFrame(cases_data)
deaths= pd.DataFrame(deaths_data)
```

```python
cases_count = cases.copy()
for i,col in enumerate(cases_count.columns):
    j = len(cases.columns)-1-i
    if j == 1:
        break
    cases_count.iloc[:,j] = cases_count.iloc[:,j] - cases_count.iloc[:,j-1]
cases_count = cases_count.set_index("STATE")
cases_count[cases_count < 0] = 0
```

```python
deaths_count = deaths.copy()
for i,col in enumerate(deaths_count.columns):
    j = len(deaths.columns)-1-i
    if j == 1:
        break
    deaths_count.iloc[:,j] = deaths_count.iloc[:,j] - deaths_count.iloc[:,j-1]
deaths_count = deaths_count.set_index("STATE")
deaths_count[deaths_count < 0] = 0
```

Finally, we want to convert the column names to pandas datetime objects so when use these subdataframes to plot the time on the x axis, it will be consistent, cleaner and visible.

```python
#Strip the 'cases' and 'deaths' at the end of each column name
cases_count.columns = cases_count.columns.str.rstrip(' cases')
deaths_count.columns = deaths_count.columns.str.rstrip(' deaths')

#Fix the name of the state column
cases_count.rename(columns={'Stat':'State'}, inplace=True)
deaths_count.rename(columns={'S':'State'}, inplace=True)
```
